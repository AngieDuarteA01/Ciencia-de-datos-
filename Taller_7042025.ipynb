```python
def ggplot_density_boxplot_colab(datos):
    sns.set_theme(style="whitegrid")
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))

    sns.kdeplot(data=datos, x='Age', hue='Survived', fill=True, alpha=0.5,
                palette=["gray", "orange"], ax=axes[0])
    sns.rugplot(data=datos, x='Age', hue='Survived', alpha=0.5,
                palette=["gray", "orange"], ax=axes[0])
    axes[0].set_title("Age")


    sns.boxplot(data=datos, x='Survived', y='Age', color='white',
                palette=["gray", "orange"], ax=axes[1])
    sns.stripplot(data=datos, x='Survived', y='Age', jitter=0.8, alpha=0.3,
                   palette=["gray", "orange"], ax=axes[1])
    axes[1].set_title("Age")

    fig.suptitle("Age", fontsize=16, y=1.02)


    axes[0].legend(title="Survived", labels=["No", "Si"], loc="upper right")
    axes[1].legend(title="Survived", labels=["No", "Si"], loc="upper right")

    plt.tight_layout()
    plt.show()


data = df_sin_nan[['Age', 'Survived']]
datos = pd.DataFrame(data)


ggplot_density_boxplot_colab(datos)

#Interpretación : Es más probable sobrevivir si tienes entre 20 y 40 años , sobretodo si tienes 35 ,pero si tienes mas de 65 años morias.
```
